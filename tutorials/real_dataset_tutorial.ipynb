# HLA_DNA_Typing Tutorial - Real Data
#### Below is an outline of how to run the HLA_DNA_Typing tool on your computer.

## 1 - Clone repository

### Get the clone link from the github repository located at https://github.com/leddavis/HLA_DNA_Typing


### Open up your terminal or command line interface and type the command: 
#### git clone /https_link_from_github/
### You will now have the HLA_DNA_Typing tool cloned into your computer.

## 2 - Install Dependencies

### The software dependencies needed for this project are listed in the tech_stack_resourses.md document within the Design_Documents folder:
### https://github.com/leddavis/HLA_DNA_Typing/blob/main/Design_Documents/tech_stack_resources.md

### The first thing you should install is an environment manager. For this tutorial, we will be using the conda manager through Anaconda3. All of the steps to download this are outlined for your convenience at this link:
### https://docs.anaconda.com/anaconda/install/index.html

### Once anaconda is installed, install the necessary packages:
- bioconda: https://bioconda.github.io/


- samtools via bioconda: https://anaconda.org/bioconda/samtools


- pandas: first use command 'conda install pip', then 


## 3 - Add sample data 
The sample data we will be using for this tutorial is in the tests/TestData folder, but you can manually place your sample data into the data folder. This is simply done on your local computer with your local files.

## 4 - Open a Python3 notebook to run the tool

[1] In your command line, navigate to the HLA_DNA_Typing project   
[2] Ensure that you are in the correct conda environment, and that the environment is activated - https://conda.io/projects/conda/en/latest/user-guide/tasks/manage-environments.html  
[3] Within your activated conda environment, open a jupyter notebook - https://jupyter-notebook-beginner-guide.readthedocs.io/en/latest/execute.html


## 5 - Import tool and data

As you can see, the sample data file used here is called "real_data_simulation.fasta" and is located in the data folder in the project. With your file, you can substitute sample_file with "data/file_name"
